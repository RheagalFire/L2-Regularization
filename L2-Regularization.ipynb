### Regularization

```python
samples,n_features= 16,1
```

```python
import numpy as np
X=np.random.random_sample(samples)* 2 * np.pi - np.pi
X=X.reshape((-1,1))
y=np.sin(X)+(np.random.random_sample(samples)*0.1).reshape((-1,1))
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.axis([-3.5,3.5,-2.33,2.33])
plt.scatter(X_train,y_train,c='g')
plt.scatter(X_test,y_test,c='purple')
plot_x=np.linspace(-np.pi,np.pi,200).reshape((-1,1))
plot_y=np.sin(plot_x)
plt.plot(plot_x,plot_y,c='r')
plt.plot()
```

```python
from sklearn.kernel_ridge import KernelRidge
```

```python
for i in range(10):
    clf = KernelRidge(degree=i, kernel='poly', alpha=0)
    clf.fit(X_train, y_train)
    plt.figure(figsize=(10,8))
    plt.axis([-3.5,3.5,-2.33,2.33])
    plt.scatter(X_train,y_train,c='g')
    plt.scatter(X_test,y_test,c='purple')
    plt.plot(np.linspace(-np.pi,np.pi,200).reshape((-1,1)),np.sin(np.linspace(-np.pi,np.pi,200).reshape((-1,1))),c='r')
    plt.plot(plot_x,clf.predict(plot_x),c='b')
    plt.plot()
```

### Regularization Formula

```python
for i in range(10):
    clf = KernelRidge(degree=9, kernel='poly', alpha=i)
    clf.fit(X_train, y_train)
    plt.figure(figsize=(10,10))
    plt.axis([-3.5,3.5,-2.33,2.33])
    plt.scatter(X_train,y_train,c='g')
    plt.scatter(X_test,y_test,c='purple')
    plt.plot(np.linspace(-np.pi,np.pi,200).reshape((-1,1)),np.sin(np.linspace(-np.pi,np.pi,200).reshape((-1,1))),c='r')
    plt.plot(plot_x,clf.predict(plot_x),c='b')
    plt.text(0,-1.5,'When alpha is={0}'.format(i))
    plt.plot()
```

Penalizing for higher slopes by using aplha making slope less steeper.

```python
x_linear=np.arange(-4,4,step=1).reshape(-1,1)
y_linear=2*x_linear+2
noise=np.array([[0],[0.5],[-1],[1],[-1],[0.5],[-0.5],[0]])
y_linear_test=y_linear+noise
plt.plot(x_linear,y_linear_test)
```

```python
for i in range(5):
    clf = KernelRidge(degree=3, kernel='poly', alpha=i)
    clf.fit(x_linear, y_linear_test)
    plt.figure(figsize=(4,4))
    plt.plot(x_linear,clf.predict(x_linear))
    plt.scatter(x_linear,y_linear_test)
    plt.plot()
```
